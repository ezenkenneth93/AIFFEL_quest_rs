```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
train = pd.read_csv("~/data/data/bike-sharing-demand/train.csv")
test = pd.read_csv("~/data/data/bike-sharing-demand/test.csv")

train.head()
```

```python
#  datetime 컬럼을 datetime 자료형으로 변환하고 

train['tempDate'] = train.datetime.apply(lambda x:x.split())
train['datetime'] = pd.to_datetime(train['datetime'])
train.head()
```

```python
train.info()
```

```python
# 연, 월, 일, 시, 분, 초까지 6가지 컬럼 생성하기
train['year'] = train.tempDate.apply(lambda x: x[0].split('-')[0])
train['month'] = train.tempDate.apply(lambda x: x[0].split('-')[1])
train['day'] = train.tempDate.apply(lambda x: x[0].split('-')[2])
train['hour'] = train.tempDate.apply(lambda x: x[1].split(':')[0])
train['minute'] = train.tempDate.apply(lambda x: x[1].split(':')[1])
train['second'] = train.tempDate.apply(lambda x: x[1].split(':')[2])

train['year'] = pd.to_numeric(train['year'], errors='coerce')
train['month'] = pd.to_numeric(train['month'], errors='coerce')
train['day'] = pd.to_numeric(train['day'], errors='coerce')
train['hour'] = pd.to_numeric(train['hour'], errors='coerce')
train['minute'] = pd.to_numeric(train['minute'], errors='coerce')
train['second'] = pd.to_numeric(train['second'], errors='coerce')

train.head()
```

```python
# sns.countplot 활용해서 시각화하기
# subplot을 활용해서 한 번에 6개의 그래프 함께 시각화하기

plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
sns.countplot(x = train.year)
# plt.title('year')

plt.subplot(2, 3, 2)
sns.countplot(x = train.month)
# plt.title('month')

plt.subplot(2, 3, 3)
sns.countplot(x = train.day)
# plt.title('day')

plt.subplot(2, 3, 4)
sns.countplot(x=train.hour)
# plt.title('hour')

plt.subplot(2, 3, 5)
sns.countplot(x = train.minute)
# plt.title('minute')

plt.subplot(2, 3, 6)
sns.countplot(x = train.second)
# plt.title('second')

plt.show()
```

```python
# X, y 컬럼 선택 및 train/test 데이터 분리
from sklearn.model_selection import train_test_split
X = train[['season', 'holiday', 'workingday', 'weather', 'temp', 'atemp', 'humidity','windspeed', 'year', 'month', 'day', 'hour', 'minute', 'second']]
y = train[['count']]

X_train, X_test, y_train, y_test = train_test_split(X, y)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```
